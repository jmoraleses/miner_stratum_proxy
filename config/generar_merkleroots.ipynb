```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import torch

# Definimos el path
path = "/content/drive/MyDrive/bitcoin2/"
df = pd.read_csv(path + "headers.csv")
```

```python
import pandas as pd

# Configurar opciones de pandas
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Suponiendo que ya tienes el DataFrame df cargado
df2 = pd.DataFrame()

# Obtener los primeros tres caracteres de merkle_root
df2['merkle_root'] = df['merkle_root'].astype(str).str[:6]

# Calcular el conteo de cada combinación
combination_counts = df2.groupby(['merkle_root']).size().reset_index(name='count')

# Filtrar las combinaciones con al menos 8 coincidencias
filtered_combination_counts = combination_counts[combination_counts['count'] >= 1]

# Ordenar los conteos en orden descendente
combination_counts_sorted = filtered_combination_counts.sort_values(by='count', ascending=False)

# Guardar los resultados en un nuevo archivo CSV
combination_counts_sorted.to_csv('merkle_counts.csv', index=False)

# Mostrar el número de combinaciones y las primeras 20 filas del resultado
print(len(combination_counts_sorted))
print(combination_counts_sorted.head(20))
```

```python
import pandas as pd

# Suponiendo que df2 ya está inicializado y cargado con las columnas 'merkle_root' y 'nonce' adecuadas
df2['merkle_root'] = df['merkle_root'].astype(str).str[:4]
df2['nonce'] = df['nonce'].astype(str).str[:4].str.zfill(4)

# Calcular el conteo de cada combinación de merkle_root y nonce
combination_counts = df2.groupby(['merkle_root', 'nonce']).size().reset_index(name='count')

# Sumar los conteos de todas las combinaciones por merkle_root
merkle_root_totals = combination_counts.groupby('merkle_root')['count'].sum().reset_index()

# Sumar los conteos de todas las combinaciones por nonce (para cada merkle_root)
nonce_totals = combination_counts.groupby(['merkle_root', 'nonce'])['count'].sum().reset_index()

# Ordenar los totales de merkle_root en orden descendente y filtrar
merkles_sorted = merkle_root_totals[merkle_root_totals['count'] >= 2].sort_values(by='count', ascending=False)

# Ordenar y filtrar totales por nonce (ajustar esta línea si es necesario para tener sentido con tu lógica)
nonces_sorted = nonce_totals[nonce_totals['count'] >= 2].sort_values(by='count', ascending=False)

# Combinar los DataFrames ordenados
# Si deseas una combinación que liste todas las entradas de merkle_roots junto con cada nonce y su conteo, puedes hacer un merge
combined_sorted = pd.merge(merkles_sorted, nonces_sorted, on='merkle_root', how='inner')

# Ver los resultados combinados
print(len(combined_sorted))
print(combined_sorted.head(20))

combined_sorted.to_csv('merkle_nonce_4-4.csv', index=False)
```

```python
import pandas as pd
import itertools

# Configurar opciones de pandas
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Suponiendo que df ya está cargado
df2 = pd.DataFrame()
# Obtener los primeros cuatro caracteres de merkle_root y nonce
df2['merkle_root'] = df['merkle_root'].astype(str).str[2]
df2['nonce'] = df['nonce'].astype(str).str[:6].str.zfill(6)

# Calcular el conteo de cada combinación
combination_counts = df2.groupby(['merkle_root', 'nonce']).size().reset_index(name='count')

# Ordenar los conteos en orden descendente
combination_counts_sorted = combination_counts[(combination_counts['count'] >= 1) & (combination_counts['count'] < 100)].sort_values(by='count', ascending=False)

# Generar todas las combinaciones posibles de nonce (4 caracteres hexadecimales)
possible_nonces = [f"{i:04x}" for i in range(16**4)]

# Crear un DataFrame vacío para almacenar los nonces faltantes
all_missing_nonces_df = pd.DataFrame()

# Iterar sobre cada merkle_root
for merkle_root in combination_counts_sorted['merkle_root'].unique():
    # Obtener los nonces existentes para el merkle_root actual
    existing_nonces = df2[df2['merkle_root'] == merkle_root]['nonce'].unique()

    # Encontrar los nonces que no están en el DataFrame para el merkle_root actual
    missing_nonces = [nonce for nonce in possible_nonces if nonce not in existing_nonces]

    # Crear un DataFrame con los nonces faltantes para el merkle_root actual
    missing_nonces_df = pd.DataFrame(missing_nonces, columns=['missing_nonce'])
    missing_nonces_df['merkle_root'] = merkle_root

    # Concatenar con el DataFrame principal
    all_missing_nonces_df = pd.concat([all_missing_nonces_df, missing_nonces_df], ignore_index=True)

# Guardar las combinaciones faltantes en un nuevo archivo CSV
all_missing_nonces_df.to_csv('missing_nonces_4-6.csv', index=False)

# Mostrar las primeras 10 filas de las combinaciones faltantes
print(len(all_missing_nonces_df))
# print(all_missing_nonces_df.head(10))
```

```python
import pandas as pd

# Configuración de Pandas para visualizar todos los datos necesarios
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Suponiendo que ya tienes el DataFrame df cargado
df2 = pd.DataFrame()

# Obtener los primeros tres caracteres de merkle_root
df2['merkle_root'] = df['merkle_root'].astype(str).str[:5]

# Calcular el conteo de cada combinación
combination_counts = df2.groupby(['merkle_root']).size().reset_index(name='count')

# Filtrar las combinaciones con al menos 8 coincidencias
filtered_combination_counts = combination_counts[combination_counts['count'] >= 8]

# Seleccionar solo los merkle_root más comunes del DataFrame original
common_merkle_roots = filtered_combination_counts['merkle_root']
filtered_df = df[df['merkle_root'].astype(str).str[:5].isin(common_merkle_roots)]

# Extraer los primeros 5 caracteres del nonce para las entradas filtradas
filtered_df['short_nonce'] = filtered_df['nonce'].astype(str).str[:5]

# Contar la frecuencia de cada nonce corto
nonce_counts = filtered_df.groupby('short_nonce').size().reset_index(name='count')

# Ordenar los conteos de nonce en orden descendente
nonce_counts_sorted = nonce_counts.sort_values(by='count', ascending=False)

# Guardar los resultados en un nuevo archivo CSV
nonce_counts_sorted.to_csv('nonce_counts_sorted.csv', index=False)

# Mostrar el número de nonces únicos y las primeras 20 filas del resultado
print(len(nonce_counts_sorted))
print(nonce_counts_sorted.head(20))
```
